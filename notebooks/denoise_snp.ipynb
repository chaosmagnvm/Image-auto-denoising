# salt and paper noise

## noised

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import dss.filters as f

pic = 28
print(pic)
img = cv2.imread(f"../noise_gauss/original_png/00{pic}.png")
img = f.snp_noise(img, p= 0.01)

rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
plt.axis("off")
```

## denoised

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import dss.filters as f
import dss.detector as d

pic = 28
print(pic)
img = cv2.imread(f"../noise_gauss/original_png/00{pic}.png")
img = f.snp_noise(img, p= 0.01)

denoised = f.denoiser_snp(img, d.snp_noise_detector(img))

rgb = cv2.cvtColor(denoised, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
plt.axis("off")
```
