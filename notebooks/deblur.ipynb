# gaussian blur

## blured

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

pic = 28
print(pic)
img = cv2.imread(f"../noise_gauss/original_png/00{pic}.png")
img = cv2.GaussianBlur(img, (0,0), 2)

rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
plt.axis("off")
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import dss.filters as f
import dss.detector as d

pic = 28
print(pic)
img = cv2.imread(f"../noise_gauss/original_png/00{pic}.png")
img = cv2.GaussianBlur(img, (0,0), 2)

deblured = f.deblur(img, d.blur_detector(img))

rgb = cv2.cvtColor(deblured, cv2.COLOR_BGR2RGB) ## problems with edges. working on it
plt.imshow(rgb)
plt.axis("off")
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import dss.filters as f
import dss.detector as d

pic = 15
print(pic)
img = cv2.imread(f"../noise_gauss/original_png/00{pic}.png")
img = cv2.GaussianBlur(img, (0,0), 2.5)

deblured = f.deblur(img, d.blur_detector(img))

rgb = cv2.cvtColor(deblured, cv2.COLOR_BGR2RGB) ## weak for strong blur
plt.imshow(rgb)
plt.axis("off")
```
